```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt # Importamos matplotlib para personalizar o gráfico

# Baixar os dados do índice Bovespa (IBOV)
# O ticker para o IBOV no Yahoo Finance é '^BVSP'
ibov_data = yf.download('^BVSP', start='2020-01-01', end='2024-08-01')

# Os exercícios pedem para usar uma "série", que é como uma única coluna de uma tabela.
# Vamos criar a série com os valores de fechamento ('Close') do IBOV.
ibov_series = ibov_data['Close']

# Vamos dar uma olhada nas primeiras linhas dos dados baixados
print("Dados do IBOV:")
print(ibov_data.head())
print("\nSérie de Fechamento (Close):")
print(ibov_series.head())
```

1 - Utilizando a série ibov_series, faça um gráfico simples que mostre a variação do valor de fechamento do índice Bovespa ao longo do tempo. Certifique-se de adicionar rótulos ao eixo X e ao eixo Y e um título ao gráfico.

**Dica:** *Utilize o método* plot() *do Pandas*

```python
# --- Resolução do Exercício 1 ---

# O Pandas se integra com o Matplotlib para criar gráficos de forma simples.
# Usamos o método .plot() diretamente na nossa série.
ibov_series.plot(
    figsize=(12, 6), # Define o tamanho da figura para melhor visualização
    title='Variação do Valor de Fechamento do IBOV (2020-2024)', # Adiciona o título do gráfico
    xlabel='Data', # Adiciona o rótulo do eixo X
    ylabel='Pontos' # Adiciona o rótulo do eixo Y
)

# Adiciona um grid para facilitar a leitura
plt.grid(True)

# Exibe o gráfico
plt.show()
```

2 - Usando a série ibov_series, calcule e exiba no console as seguintes estatísticas básicas: valor máximo, valor mínimo, média e mediana do índice Bovespa no período analisado.

**Dicas**: *Utilize os métodos* max(), min(), mean() e median() *da série Pandas para calcular as estatísticas. Experimente exibir os resultados de forma clara e organizada.*

```python
# --- Resolução Final e à Prova de Futuro (Exercício 2) ---

# 1. Garantir que estamos trabalhando com a série de fechamento
ibov_series = ibov_data['Close']

# 2. Calcular as estatísticas.
#    Como o seu ambiente retorna uma Série, usamos .iloc[0] para extrair o valor numérico.
#    Esta é a forma correta e recomendada pelo aviso.
valor_maximo = ibov_series.max().iloc[0]
valor_minimo = ibov_series.min().iloc[0]
media = ibov_series.mean().iloc[0]
mediana = ibov_series.median().iloc[0]

# 3. Exibir os resultados.
#    Agora as variáveis já são números e a impressão funcionará sem erros ou avisos.
print("--- Estatísticas Básicas do IBOV (2020-2024) ---")
print(f"Valor Máximo (pico): {valor_maximo:.2f} pontos")
print(f"Valor Mínimo (vale): {valor_minimo:.2f} pontos")
print(f"Média do Período:    {media:.2f} pontos")
print(f"Mediana do Período:   {mediana:.2f} pontos")
```
